```python
import numpy as np
from neural_network import ThreeLayerNN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

class Generator:
    def __init__(self, z_dim=10, hidden_units=8, data_dim=10, learning_rate=0.0002, clip_value=1.0):
        self.z_dim = z_dim
        self.hidden_units = hidden_units
        self.data_dim = data_dim
        self.learning_rate = learning_rate
        self.clip_value = clip_value

        self.W1 = np.random.randn(self.z_dim, self.hidden_units) * np.sqrt(2.0 / z_dim)
        self.b1 = np.zeros((1, self.hidden_units))

        self.W2 = np.random.randn(self.hidden_units, self.data_dim) * np.sqrt(2.0 / hidden_units)
        self.b2 = np.zeros((1, self.data_dim))

    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return np.where(x > 0, 1, 0)
    
    def batch_norm(self, x, eps=1e-5):
        mean = np.mean(x, axis=0, keepdims=True)
        var = np.var(x, axis=0, keepdims=True)
        return (x - mean) / np.sqrt(var + eps)

    def forward(self, z):
        self.z_input = z
        
        self.z1 = np.dot(z, self.W1) + self.b1
        self.z1_bn = self.batch_norm(self.z1)
        self.a1 = self.relu(self.z1)

        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.z2_bn = self.batch_norm(self.z2)
        self.output = self.z2
        return self.output

    def backward(self, grad_output):
        dZ2 = grad_output
        dZ2 = np.clip(dZ2, -self.clip_value, self.clip_value)
        dW2 = np.dot(self.a1.T, dZ2)
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = np.dot(dZ2, self.W2.T)
        dZ1 = dA1 * self.relu_derivative(self.z1)

        dZ1 = np.clip(dZ1, -self.clip_value, self.clip_value)

        dW1 = np.dot(self.z_input.T, dZ1)
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        dW2 = np.clip(dW2, -self.clip_value, self.clip_value)
        db2 = np.clip(db2, -self.clip_value, self.clip_value)
        dW1 = np.clip(dW1, -self.clip_value, self.clip_value)
        db1 = np.clip(db1, -self.clip_value, self.clip_value)

        self.W2 -= self.learning_rate * dW2
        self.b2 -= self.learning_rate * db2
        self.W1 -= self.learning_rate * dW1
        self.b1 -= self.learning_rate * db1

z_dim = 10
input_dim = 10
hidden_units = 8
learning_rate = 0.0002

decay_rate = 0.99
decay_steps = 1000
epochs = 100000
batch_size = 64

real_data, _ = make_blobs(n_samples=1000, n_features=input_dim, centers=2, cluster_std=1.0, random_state=42)

discriminator = ThreeLayerNN(
    input_dim = input_dim,
    hidden_units = hidden_units,
    output_dim = 1,
    learning_rate = learning_rate,
    clip_value=1.0
)

generator = Generator(
    z_dim = z_dim,
    hidden_units = hidden_units,
    data_dim = input_dim,
    learning_rate = learning_rate,
    clip_value=1.0
)

num_samples = 1000
# real_data = np.random.normal(0, 1, size=(num_samples, input_dim))

def plot_data(gen, z_dim, num_samples=100):
    z_vis = np.random.normal(0, 1, size=(num_samples, z_dim))
    samples = gen.forward(z_vis)

    if samples.shape[1] >= 2:
        plt.figure(figsize=(6, 6))
        plt.scatter(real_data[:, 0], real_data[:, 1], alpha=0.6, label='Real Data')
        plt.scatter(samples[:, 0], samples[:, 1], alpha=0.6, label='Generated Data')
        plt.title(f"Generated Data at Epoch {epoch}")
        plt.xlabel("Dimension 1")
        plt.ylabel("Dimension 2")
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Generated samples (first 5 rows):")
        print(samples[:5])

for epoch in range(epochs):

    current_lr = learning_rate * (decay_rate ** (epoch / decay_steps))
    discriminator.learning_rate = current_lr
    generator.learning_rate = current_lr

    idx = np.random.randint(0, real_data.shape[0], batch_size)
    real_batch = real_data[idx]

    z = np.random.normal(0, 1, size=(batch_size, z_dim))

    fake_batch = generator.forward(z)

    y_real = np.ones((batch_size, 1))
    y_fake = np.zeros((batch_size, 1))

    d_real = discriminator.forward(real_batch)
    d_real = np.clip(d_real, 1e-7, 1 - 1e-7) 
    discriminator.backward(real_batch, y_real, d_real)

    d_fake = discriminator.forward(fake_batch)
    d_fake = np.clip(d_fake, 1e-7, 1 - 1e-7)
    discriminator.backward(fake_batch, y_fake, d_fake)

    fake_batch_for_g = generator.forward(z)
    d_fake_for_g = discriminator.forward(fake_batch_for_g)
    d_fake_for_g = np.clip(d_fake_for_g, 1e-7, 1 - 1e-7)

    eps = 1e-15
    grad_outputs = -1.0 / d_fake_for_g

    grad_outputs_expanded = np.repeat(grad_outputs, input_dim, axis=1)

    generator.backward(grad_outputs_expanded)

    if epoch % 100 == 0:
        
        d_loss_real = np.mean(np.log(d_real + eps))
        d_loss_fake = np.mean(np.log(1 - d_fake + eps))
        d_loss = np.mean(d_loss_real + d_loss_fake)

        g_loss = np.mean(np.log(d_fake_for_g + eps))

        print(f"Epoch: {epoch}, D loss: {d_loss:.4f}, G loss: {g_loss:.4f}")

        if epoch % 500 == 0:
            plot_data(generator, z_dim, num_samples=5)
```
